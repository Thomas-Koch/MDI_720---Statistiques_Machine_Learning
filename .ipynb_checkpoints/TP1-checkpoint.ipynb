Le mot “régression” a été introduit par Sir Francis Galton (cousin de C. Darwin) alors qu’il
étudiait la taille des individus au sein d’une descendance. Il tentait de comprendre pourquoi
les grands individus d’une population semblaient avoir des enfants d’une taille plus petite, plus
proche de la taille moyenne de la population ; d’où l’introduction du terme “régression”. Dans la
suite on va s’intéresser aux données récoltées par Galton.

Q1 Récupérer les données du fichier https://bitbucket.org/portierf/shared_files/downloads/
Galton.txt (voir aussi leur description ici http://www.randomservices.org/random/
data/Galton.html) et charger les avec Pandas.

```python
import pandas as pd
import numpy as np

# mon module standard de lecture de fichier via pandas
na_values = ['?', ''] # qui viennent s'ajouter aux autres valeurs nulles
fields = ['Family', 'Father', 'Mother', 'Gender', 'Height', 'Kids'] # on se limite à ces champs
df = pd.read_csv("./Galton.txt",sep='\t',nrows=200000, na_values=na_values, usecols=fields) # séparateur tabulation

print("ok") # pour annoncer la fin du chargement...
```

Q2 Créer une colonne supplémentaire appelée “MeanParents” qui contient la taille du parent
“moyen”, c’est-à-dire $\frac{1}{2}$
$*$ $(taille(père) + 1.08 * taille(mère))$

```python
df['MeanParents'] = (df['Father'] + 1.08 * df['Mother']) / 2
```

On note $x_i$ la taille du parent moyen pour la famille i et $y_i$ la taille de l’enfant. On écrit
$y_i = \theta_1  x_i + \theta_0 + \epsilon_i$ et on modélise les variables $\epsilon_i$ comme centrées, indépendantes de même variance $\sigma^2$ inconnue.

Q3 Tracer le nuage de points ($x_i, y_i$) pour $1 \leq i \leq n$ où n est le nombre d’observations figurant
dans les données.

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7, 7))
plt.plot(df['MeanParents'],df['Height'], 'g+')
plt.xlabel('Parent moyen')
plt.ylabel('Enfant')
plt.show()
```

Q4 Estimer $\theta_0, \theta_1$ en utilisant la fonction LinearRegression de sklearn. Calculer
et visualiser les valeurs prédites $y_i = \theta_1  x_i + \theta_0$ et $y_i$ sur un même graphique.

```python
import sklearn.linear_model as lm
import statsmodels.api as sm

y = df['Height']
X = df[['MeanParents']] # [[ ]]pour avoir un Dataframe pour utiliser LinearRegression

# Modèle 1 : avec intercept
modele_1 = lm.LinearRegression(fit_intercept=True) 
modele_1.fit(X, y) # On entraine le modèle

y_pred_modele_1 = modele_1.predict(X)# Prédiction 

print("theta0(intercept) =", round(modele_1.intercept_, 3))
print("theta1(pente) =", round(modele_1.coef_[0],3))
```

```python
fig = plt.figure(figsize=(7, 7))
plt.plot(df['MeanParents'], y, 'g+', label="Data")
plt.plot(df['MeanParents'], y_pred_modele_1,
         label="Régression linéaire avec intercept",color='g')
plt.legend(loc='upper left')
plt.xlabel('Parent moyen')
plt.ylabel('Enfant')
plt.show()
```

Q5 Vérifier la formule vue en cours liée au recentrage des données. On centrera les données et
on vérifiera que la prédiction effectuée dans le modèle centré est bien la même que celle
effectuée précédemment. Justifiez votre réponse.

```python
# on va calculer une nouvelle prédiction à partir des données recentrées et vérifier que les vecteurs
# de prédiction calculés par les 2 méthodes sont égaux (allclose)  

# on ajoute des colonnes au dataframe avec les données centrées suffixées par _c
df['Height_c'] = df['Height']-df['Height'].mean()
df['MeanParents_c'] = df['MeanParents']-df['MeanParents'].mean()

y_c = df['Height_c']
X_c = df[['MeanParents_c']] 

# Modèle 2 : sans intercept
modele_2 = lm.LinearRegression(fit_intercept=False) 
modele_2.fit(X_c, y_c) # On entraine le modèle

y_pred_modele_2 = modele_2.predict(X_c) # Prédiction

# on vérifie que la pente est la même que précédemment
print("Avec sklearn ->")
print("theta0(intercept) =", round(modele_2.intercept_))
print("theta1(pente) =", round(modele_2.coef_[0],3))

# calcul avec les formules du cours
print()
print("Avec les formules du cours ->")
print("theta0(intercept) = 0")
theta_1_calcul = (y_c * df['MeanParents_c'] ).sum() / (
                        (df['MeanParents_c'] * df['MeanParents_c']).sum())
print("theta1(pente) =", round(theta_1_calcul,3))

# On vérifie que les vecteurs de prédiction calculés par les 2 méthodes sont égaux (allclose)
# après la translation de la taille moyenne sur y centré
print()
print("Les prédictions sont-elles quasi égales ->")
print("Egalité des vecteurs de prédiction ?", \
    np.allclose(y_pred_modele_1, y_pred_modele_2 + df['Height'].mean()))
print("Différence entre les pentes calculées :","{:.2E}".format(theta_1_calcul- modele_2.coef_[0]))
```

 $\color{blue}{\text{La pente est identique à celle du premier modèle et les 2 vecteurs de prédiction sont quasi égaux.}}$

Q9 A l’aide d’un partitionnement aléatoire de l’échantillon, construire un échantillon de test et un échantillon d’apprentissage. L’échantillon test doit contenir 20% des données. On estimera, sur l’échantillon d’apprentissage deux modèles : un premier modèle basé uniquement sur les garçons, et un deuxième modèle basé sur l’échantillon complet.
On comparera les deux prédicteurs sur le groupe “garçon” de l’échantillon test en calculant
le risque de prédiction :          $\frac{1}{|test|}\sum_{i \in test}^{} (y_i - \hat{y}_i)^2$


```python
 from sklearn.model_selection import train_test_split
    
# On garde une partie données pour le test (20%) et le reste (80%) sert à 
# l'entrainement du modèle
df_train, df_test = train_test_split(df, test_size=0.2)

y_train = df_train['Height']
X_train = df_train[['MeanParents']]

#De ces échantillons, on extraie les garçons
y_garcon_train = df_train[df_train['Gender']=='M']['Height']
X_garcon_train = df_train[df_train['Gender']=='M'][['MeanParents']]

y_garcon_test = df_test[df_test['Gender']=='M']['Height']
X_garcon_test = df_test[df_test['Gender']=='M'][['MeanParents']]
```

```python
modele_garcon.fit(X_garcon_train, y_garcon_train) # On entraine le modèle
y_pred_modele_garcon = modele_garcon.predict(X_garcon_test) # Prédiction

modele_1.fit(X_train, y_train) # On entraine le modèle
y_pred_modele_1 = modele_1.predict(X_garcon_test) # Prédiction
```

Calcul du risque de prédiction des 2 modèles

```python
n_garcon = y_garcon_test.count() # taille de l'échantillon d'entrainement des garçons

# Calcul des risques de prédiction des 2 modèles sur les garçons
risque_modele_garcon = ((y_garcon_test - y_pred_modele_garcon) * (y_garcon_test - y_pred_modele_garcon)).sum() / n_garcon
risque_modele_1 = ((y_garcon_test - y_pred_modele_1) * (y_garcon_test - y_pred_modele_1)).sum() / n_garcon

print("Sur les garçons ->")
print("Risque de prédiction du modèle Garçon :", round(risque_modele_garcon,2))
print("Risque de prédiction du modèle 1 (Mixte) :", round(risque_modele_1,2))
```

Q10 En combinant les deux prédicteurs, celui appris sur les filles et celui appris sur les garçons,
calculer le risque de prédiction associé. Pour chaque indicidu “garçon” on applique le modèle
appris sur les garçon, pour chaque individu “fille” on lui applique le modèle appris sur les
filles. On calculera ensuite le risque prédictif.

```python
#Des échantillons train et test, on extraie les filles
y_fille_train = df_train[df_train['Gender']=='F']['Height']
X_fille_train = df_train[df_train['Gender']=='F'][['MeanParents']]

y_fille_test = df_test[df_test['Gender']=='F']['Height']
X_fille_test = df_test[df_test['Gender']=='F'][['MeanParents']]

# Régression sur les filles 
modele_fille = lm.LinearRegression(fit_intercept=True)
modele_fille.fit(X_fille_train, y_fille_train) # On entraine le modèle
y_pred_modele_fille = modele_fille.predict(X_fille_test) # Prédiction
```

```python
n_fille = y_fille_test.count() # taille de l'échantillon d'entrainement des filles

# Calcul du risque de prédiction du modèle Fille sur les filles
risque_modele_fille = ((y_fille_test - y_pred_modele_fille) * (y_fille_test - y_pred_modele_fille)).sum() / n_fille

print ("Sur les filles ->")
print ("Risque de prédiction du modèle Fille :", round(risque_modele_fille,2))
```

```python
# Calcul du risque de prédiction global en appliquant le modèle dédié à chaque sexe
risque_global = (((y_garcon_test - y_pred_modele_garcon) * (y_garcon_test - y_pred_modele_garcon)).sum() +
                    ((y_fille_test - y_pred_modele_fille) * (y_fille_test - y_pred_modele_fille)).sum()
                      ) / (n_garcon + n_fille)

print ("Sur la population mixte ->")
print ("Risque de prédiction du modèle global :", round(risque_global,2))


```

Q11 En utilisant, un encodage booléen de la variable GENDER construire le même prédicteur que précédement. Calculer, "à la main", son coefficient de determination. On comparera ce dernier au coefficient de determination du modèle avec seulement la variable MeanParents. On pourra se référer au polycopié pour la définition du coefficient de determination (dit aussi le R2).

```python
y_test = df_test['Height']
X_test = df_test[['MeanParents']]
# Prédiction combinée sur les 2 sexes (les 2 modèles sont déjà entrainés sur le jeu de "train")
y_pred_2sexes = (df_test['Gender'] == 'M') * modele_garcon.predict(X_test) + \
                (df_test['Gender'] == 'F') * modele_fille.predict(X_test)

# calcul du risque de prédiction 
n = y_test.count() 
risque_modele_2sexes = ((y_test - y_pred_2sexes) * (y_test - y_pred_2sexes)).sum() / n
print ("Risque de prédiction du modèle '2 sexes' :", round(risque_modele_2sexes,2))

# calcul du coefficient de détermination
y_mean =df_test['Height'].mean()
R2_modele_2sexes = ((y_pred_2sexes - y_mean) * (y_pred_2sexes - y_mean)).sum() /  \
    ((y_test - y_mean) * (y_test - y_mean)).sum()
print ("Coefficient de détermination du modèle '2 sexes' :", round(R2_modele_2sexes,2))

# Calcul le modèle 1 sur la population de test
y_pred_modele_1 = modele_1.predict(X_test)# Prédiction

# calcul du coefficient de détermination
R2_modele_1 = ((y_pred_modele_1 - y_mean) * (y_pred_modele_1 - y_mean)).sum() /  \
    ((y_test - y_mean) * (y_test - y_mean)).sum()
print ("Coefficient de détermination du modèle 1:", round(R2_modele_1,2))
```

$\color{blue}{\text{Le coefficient de détermination (R2) du modèle '2 sexes' est beaucoup plus proche de 1 que celui de modèle 1.}}$
$\color{blue}{\text{Le modèle '2 sexes' fournit une prédiction plus proche de la réalité observée.}}$

Q12 Faire une régression avec les 2 variables explicatives suivantes : la taille du père et de la
mère. Faire un graphique (en 3 dimensions) qui représente le nuage de point et le plan de
régression. Visualiser la distribution des résidus.

```python
from mpl_toolkits import mplot3d
from matplotlib import cm

y = df['Height']
X = df[['Father','Mother']]

# Régression sur la taille sur père et de la mère 
modele_pere_mere = lm.LinearRegression(fit_intercept=True)
modele_pere_mere.fit(X, y) # On entraine le modèle
y_pred = modele_pere_mere.predict(X) # Prédiction

# figure en 3D
fig = plt.figure(figsize=(13, 9))
ax = plt.axes(projection='3d')

# nuage de points
ax.plot3D(df['Father'],df['Mother'],df['Height'], 'g+')

# plan de régression
X = np.arange(df['Father'].min(), df['Father'].max(), 1)
Y = np.arange(df['Mother'].min(), df['Mother'].max(), 1)
X, Y = np.meshgrid(X, Y)
Z=modele_pere_mere.coef_[0]*X + modele_pere_mere.coef_[1]*Y + modele_pere_mere.intercept_
ax.plot_wireframe(X,Y,Z,rstride=2, cstride=2)
ax.set_xlabel('Père')
ax.set_ylabel('Mère')
ax.set_zlabel('Enfant')
plt.show()
```

$\color{blue}{\text{L'orientation du plan de régression montre que plus le père et / ou la mère est grand(e), plus l'enfant a des chances d'être grand.}}$

```python
residus = y - y_pred

# on dessine l'histogramme
n, bins, patches = plt.hist(residus, 40, density=1, facecolor='g')
plt.xlabel('Residus')
plt.ylabel('Probability')
plt.title('Histogramme des résidus')
plt.grid(True)
plt.show() 
```

Q13 A l’aide d’un échantillon test (construit comme précédemment), comparer la performance
du modèle précédent à celle du modèle contenant seulement la variable MeanParents. Commenter.

```python
# On relance un échantillonage, avec pour le test (20%) et le reste (80%) sert à 
# l'entrainement du modèle
df_train, df_test = train_test_split(df, test_size=0.2)

y_train = df_train['Height']
X_train = df_train[['MeanParents']] 
X_train_pere_mere  = df_train[['Father','Mother']]
y_test = df_test['Height']
X_test = df_test[['MeanParents']] 
X_test_pere_mere  = df_test[['Father','Mother']]

# Entrainement et prédiction des 2 modèles 
modele_1.fit(X_train, y_train) # On entraine le modèle
modele_pere_mere.fit(X_train_pere_mere, y_train) # On entraine le modèle
y_pred_modele_1 = modele_1.predict(X_test) # Prédiction
y_pred_pere_mere = modele_pere_mere.predict(X_test_pere_mere) # Prédiction

# on calcule les risques de prédiction des 2 modèles
n = y_test.count() 
risque_modele_1 = ((y_test - y_pred_modele_1) * (y_test - y_pred_modele_1)).sum() / n
risque_modele_pere_mere = ((y_test - y_pred_pere_mere) * (y_test - y_pred_pere_mere)).sum() / n
print ("Risque de prédiction du modèle parent moyen :", round(risque_modele_1,2))
print ("Risque de prédiction du modèle père et mère :", round(risque_modele_pere_mere,2))
```

$\color{blue}{\text{Les résultats des 2 modèles sont très proches. Ce n'est pas surprenant, ce qui discrimine c'est le sexe de l'enfant,}}$
$\color{blue}{\text{le fait de séparer les variables père et mère ne résoud pas l'hétérogénité des tailles selon le sexe de l'enfant.}}$
